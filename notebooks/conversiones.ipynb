```python
columnas = ['FECHA', 'COD C.', 'MARCA / MODELO', 'PLACA', 'TANQUE', 'CAPACIDAD', 'UBICACIÓN', 'P.TANQUE', 'KIT', 'K. MARCA', 'P. KIT', 'CERTIFICADO', 'P. CERT', 'P. CINTA NEGRA', 'P. PRESINTOS', 'P. MANGUERA', 'PV. Cliente', 'EFECTIVO', 'BCP', 'TARJETA']
# df_conversiones = pd.DataFrame(columns=columnas)

# Limpiamos y convertimos los nombres de las columnas
df_conversiones.columns = [col.lower().strip() for col in df_conversiones.columns]
df_conversiones.columns
```

```python
df_conversiones
```

```python
df_conversiones = df_conversiones.copy()

df_conversiones['ganancia'] = df_conversiones['pv. cliente'] - df_conversiones['costo']

totales = {
    "ventas_totales": df_conversiones["pv. cliente"].sum(),
    "total_ganancia": df_conversiones["ganancia"].sum(),
    "total_retorno_capital": df_conversiones["costo"].sum(),
    "total_efectivo": df_conversiones["efectivo"].sum(),
    "total_bcp": df_conversiones["bcp"].sum(),
    "total_tarjeta": df_conversiones["tarjeta"].sum()
}

totales
```

```python
# Calculando la proporción de la ganancia sobre el precio de venta para cada venta
df_conversiones["proporcion_ganancia"] = df_conversiones["ganancia"] / df_conversiones["pv. cliente"]

# Calculando cuánto de cada método de pago corresponde a ganancia
df_conversiones["ganancia_efectivo"] = df_conversiones["efectivo"] * df_conversiones["proporcion_ganancia"]
df_conversiones["ganancia_bcp"] = df_conversiones["bcp"] * df_conversiones["proporcion_ganancia"]
df_conversiones["ganancia_tarjeta"] = df_conversiones["tarjeta"] * df_conversiones["proporcion_ganancia"]

# Calculando cuánto de cada método de pago corresponde al retorno del capital
df_conversiones["retorno_efectivo"] = df_conversiones["efectivo"] - df_conversiones["ganancia_efectivo"]
df_conversiones["retorno_bcp"] = df_conversiones["bcp"] - df_conversiones["ganancia_bcp"]
df_conversiones["retorno_tarjeta"] = df_conversiones["tarjeta"] - df_conversiones["ganancia_tarjeta"]

# Sumando totales para cada método de pago en cuanto a ganancia y retorno de capital
totales_metodos_pago = {
    "ganancia_total_efectivo": round(df_conversiones["ganancia_efectivo"].sum(),2),
    "retorno_total_efectivo": round(df_conversiones["retorno_efectivo"].sum(),2),
    "ganancia_total_bcp": round(df_conversiones["ganancia_bcp"].sum(),2),
    "retorno_total_bcp": round(df_conversiones["retorno_bcp"].sum(),2),
    "ganancia_total_tarjeta": round(df_conversiones["ganancia_tarjeta"].sum(),2),
    "retorno_total_tarjeta": round(df_conversiones["retorno_tarjeta"].sum(),2),
}

totales_metodos_pago
```
